```python
import os
import numpy as np
from OpenGL.GL import glLineWidth
import pyqtgraph as pg
import pyqtgraph.opengl as gl
import sys
```

```python
class kitti_object(object):
    '''Load and parse object data into a usable format.'''    
    def __init__(self, root_dir, split="velodyne_frompcd"):
        '''root_dir contains training and testing folders'''
        self.root_dir = root_dir
        self.split = split
        self.split_dir = os.path.join(root_dir, split)
        #
        self.lidar_dir = os.path.join('../velodyne_frompcd', self.split_dir)
        self.label_dir = os.path.join('label', self.split_dir)
    def get_lidar(self, idx): 
        lidar_filename = os.path.join(self.lidar_dir, '%06d.bin'%(idx))
        return load_velo_scan(lidar_filename)
```

```python
def load_velo_scan(velo_filename):
    scan = np.fromfile(velo_filename, dtype=np.float32)
    scan = scan.reshape((-1, 4))
    print(scan)
    return scan
```

```python
#algorithm to edit from new dataset to shift x +20, z plus 1 and normalize the intensity
dataset = kitti_object('/home/yonas/bin/5thbatch')
for x in range(200):
    data_idx = x
    lidar_data = dataset.get_lidar(data_idx)
    lidar_data_edited=lidar_data 
    lidar_data_edited[:,3]=lidar_data_edited[:,3]/np.max(lidar_data[:,3]) #normalize intensity
    lidar_data_edited[:,0]=lidar_data_edited[:,0]+20 
    lidar_data_edited[:,2]=lidar_data_edited[:,2]-1 
    lidar_data_edited.tofile(str(x)+'.bin')
```

```python
# Rotation matrix of A with respect to B.
R_A_B = np.array(((1,0,0),(0,0,-1),(0,1,0)))    # Get 3x3 matrix
# The translation is the origin of A in B.
tAorg_B = np.array([[1,2,4]]).T   # Get as a 3x1 matrix
# H_A_B means transform A to B.
H_A_B = np.block([[R_A_B, tAorg_B], [0,0,0,1]])  # Get 4x4 matrix
# Define a point in the A frame, as [x,y,z,1].
P_A = np.array([[-1,0,1,1]]).T    # Get as a 4x1 matrix
# Convert point to B frame.
P_B = H_A_B @ P_A
P_B
```
